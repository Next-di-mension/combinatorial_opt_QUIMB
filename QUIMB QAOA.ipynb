Creating an ansatz circuit 

In the tensor network, the tensors represent the quantum gates which are replicating the unitary evolution in hand. 

$U3$ gate is written as 


```python
def single_qubit_layer(circ, gate_round=None):
    """Apply a parametrizable layer of single qubit ``U3`` gates.
    """
    for i in range(circ.N): # N qubits 
        # initialize with random parameters
        params = qu.randn(3, dist='uniform') #Fast multithreaded generation of random normally distributed data
        # shape of the output random array = 3, dist: type of random number to generate
        # as U3 has 3 paameters (theta, phi and lambda) 
        circ.apply_gate(
            'U3', *params, i, 
            gate_round=gate_round, parametrize=True)
```

```python
def two_qubit_layer(circ, gate2='CZ', reverse=False, gate_round=None):
    """Apply a layer of constant entangling gates.
    """
    regs = range(0, circ.N - 1)
    if reverse:
        regs = reversed(regs)
    
    for i in regs:
        circ.apply_gate(
            gate2, i, i + 1, gate_round=gate_round)
```

```python
def ansatz_circuit(n, depth, gate2='CZ', **kwargs):
    """Construct a circuit of single qubit and entangling layers.
    """
    circ = qtn.Circuit(n, **kwargs) #Class for simulating quantum circuits using tensor networks.
    # n = number of qubits 
    for r in range(depth):
        # single qubit gate layer
        single_qubit_layer(circ, gate_round=r)
        
        # alternate between forward and backward CZ layers
        two_qubit_layer(
            circ, gate2=gate2, gate_round=r, reverse=r % 2 == 0)
        
    # add a final single qubit layer
    single_qubit_layer(circ, gate_round=r + 1)
    
    return circ
```

```python
def energy(x):
    p = len(x) // 2
    gammas = x[:p]
    betas = x[p:]
    circ = qtn.circ_qaoa(terms, p, gammas, betas)

    ZZ = qu.pauli('Z') & qu.pauli('Z')

    ens = [
        circ.local_expectation(weight * ZZ, edge, optimize=opt, backend="jax") # Compute the a single expectation value of operator G, acting on sites where, making use of reverse lightcone cancellation.
        for edge, weight in terms.items()
    ]
    
    return sum(ens).real
```
